# Computing continuous cooling diagram from time-temperature transformation diagram

## Program to construct a TTT diagram theoretically by computational approach

```python
#importing the required module
#-----------------------------
import math
import array as arr
from numpy import *
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Working Temperature Range
#-------------------------
Temp=list(range(725,299,-1))
print(Temp)
```

```python
#loop to calculate avrami parameters 'k' and 'n'
#----------------------------------------------
calc_k=[]
for i in Temp:
    x=0.0000751+0.022*math.exp(-0.5*((i-536)/40.81)**2.0)
    calc_k.append(x)
print(np.round_(calc_k, decimals = 4))

calc_n=[]
for i in Temp:
    x=1.7852+1.54841*math.exp(-0.5*((i-559.82)/60.866)**2)
    calc_n.append(x)
print(np.round_(calc_n, decimals = 4))
```

```python
#loop to calculate transformation time 't'
#----------------------------------------
time_1=[]
for i, j in zip(calc_k, calc_n):
        x=((-0.010050335)/(-i))**(1/j)
        time_1.append(x)        
print(np.round_(time_1, decimals = 5))

time_50=[]
for i, j in zip(calc_k, calc_n):
        x=((-0.69314718)/(-i))**(1/j)
        time_50.append(x)        
print(np.round_(time_50, decimals = 5))

time_99=[]
for i, j in zip(calc_k, calc_n):
        x=((-4.605170186)/(-i))**(1/j)
        time_99.append(x)        
print(np.round_(time_99, decimals = 5))
```

```python
#Plotting Theoretical TTT Diagram
#--------------------------------
t = Temp
a = time_1
b = time_50
c = time_99

plt.plot(a,t, 'r') # plotting t, a separately 
plt.plot(b,t, 'b') # plotting t, b separately 
plt.plot(c,t, 'g') # plotting t, c separately 

#convert x-axis to Logarithmic scale
#-----------------------------------
plt.xscale("log")

#naming the x and y axis
#-----------------------
plt.xlabel('time')
plt.ylabel('temp')
 
#giving a title and show graph
#-----------------------------
plt.title('TTT Diagrm')
plt.show()
```
